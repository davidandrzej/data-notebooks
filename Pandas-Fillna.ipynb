```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
data = pd.Series(np.random.randn(10))
data.plot()
plt.show()
```

```python
data[5:7] = None
data.plot()
plt.show()
```

```python
data.fillna(method='backfill').plot()
plt.show()
```

```python
data.fillna(method='ffill').plot()
plt.show()
```

```python
data.fillna(data.mean()).plot()
plt.show()
```

```python
data.interpolate().plot()
plt.show()
```
